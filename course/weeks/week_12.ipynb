# שבוע 12 — Pandas: עבודה על טבלאות

בשבוע האחד עשר נכיר את **Pandas**, הכלי המרכזי לעיבוד וניתוח נתונים.

נלמד את עקרונות העבודה עם **Pandas** – איך לחשוב על נתונים בצורה טבלאית,  
מהם מבני הנתונים המרכזיים שלה, ואיך לייבא נתונים ממקורות שונים לקובצי DataFrame.

בסוף השבוע תדעו:
- ליצור ולעבוד עם מבני הנתונים `Series` ו־`DataFrame`  
- לייבא ולייצא נתונים 
- לבצע פעולות בסיסיות על טבלאות כמו גישה לשורות ועמודות, הוספה ומחיקה של תוכן מטבלה
