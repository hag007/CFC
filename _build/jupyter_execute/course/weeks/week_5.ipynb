# שבוע 5 — זיכרון, פונקציות מתקדמות ומבנים בלתי־משתנים

בשבוע החמישי נעמיק אל מאחורי הקלעים של פייתון ונבין כיצד היא **שומרת מידע בזיכרון**.  
נלמד על ההבדל החשוב בין אובייקטים *ניתנים לשינוי* (mutable) לבין *בלתי־ניתנים לשינוי* (immutable),  
ונראה כיצד זה משפיע על משתנים, על העתקות, ועל התנהגות פונקציות.

נמשיך לעסוק ביכולות מתקדמות של פונקציות — העברת פרמטרים, ערכים חוזרים, והקצאת זיכרון,  
ונראה כיצד לכתוב קוד יעיל וברור שמתחשב באופן שבו פייתון מנהלת את האובייקטים שלנו בזיכרון.

בסוף השבוע תדעו:
- להבחין בין אובייקטים ניתנים ובלתי־ניתנים לשינוי  
- להבין כיצד מועברים משתנים לפונקציות ומה המשמעות של שינוי ערכים בתוך פונקציה  
- להשתמש בפונקציות בצורה מתקדמת ויעילה  
- לזהות מצבים בהם נדרש להעתיק או לשכפל אובייקטים כדי למנוע שגיאות
