# שבוע 10 — חישובים מדעיים עם SciPy ו־מבוא לעבודה עם Pandas

בשבוע העשירי נשלב בין עולמות — נצלול אל הספרייה **SciPy**, אחת מאבני היסוד של מדעי הנתונים והחישוב הפיזיקלי בפייתון,  
ונתחיל גם את ההיכרות עם **Pandas**, הכלי המרכזי לעיבוד וניתוח נתונים.

**חלק א: חישובים מדעיים עם SciPy**

נלמד כיצד לבצע אינטגרציה, אופטימיזציה, אינטרפולציה, ולעבוד עם קבועים פיזיקליים –  
כלים קריטיים לכל פיזיקאי, מהנדס או חוקר העוסק בנתונים אמפיריים ובמודלים מתמטיים.

נבין כיצד ניתן להשתמש ב־SciPy כדי:
- לגשת לקבועים פיזיקליים מדויקים כמו $c, G, h$ ולתרגם יחידות בין מערכות שונות  
- לחשב אינטגרלים של פונקציות (חד־ממדיים ורב־ממדיים) בצורה נומרית  
- למצוא מינימום ומקסימום של פונקציות באמצעות שיטות אופטימיזציה מתקדמות  

**חלק ב: מבוא לעבודה עם Pandas**

נלמד את עקרונות העבודה עם **Pandas** – איך לחשוב על נתונים בצורה טבלאית,  
מהם מבני הנתונים המרכזיים שלה, ואיך לייבא נתונים ממקורות שונים לקובצי DataFrame.

בסוף השבוע תדעו:
- להשתמש במודול `scipy.constants` לקבועים ויחידות פיזיקליות  
- לבצע אינטגרציה נומרית עם `scipy.integrate`  
- לפתור בעיות אופטימיזציה עם `scipy.optimize`  
- ליצור ולעבוד עם מבני הנתונים `Series` ו־`DataFrame`  
- לייבא ולייצא נתונים מקבצים שונים (CSV, Excel ועוד)
